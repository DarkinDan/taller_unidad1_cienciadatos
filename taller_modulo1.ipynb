### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí
# Promedio
suma = 0
cont = 0
maximo = esperanzas_de_vida[0]
minimo = esperanzas_de_vida[0]

for numero in esperanzas_de_vida:
    suma += numero
    cont +=1

    if numero > maximo: # analizar numero maximo
        maximo = numero

    if numero < minimo: # analizar numero minimo
        minimo = numero
promedio = suma / cont
print(f"El promedio de la lista 'esperanzas_de_vida' es {promedio}")
  
# Máximo
print(f"El numero mayor de la lista 'esperanzas_de_vida' es {maximo}")

# Mínimo
print(f"El mumero menor de la lista 'esperanzas_de_vida' es {minimo}")

````

## Respuesta:
- `Promedio =`  70.21000000000001
- `Maximo =` 83.2
- `Minimo =` 49.3

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí
def clasificar_esperanza_vida(years:float)->str:
    if years >= 75:
        return "Alta"
    elif years >= 60 and years <= 74.9:
        return "Media"
    elif years < 60:
        return "Baja"
    return ""

print(clasificar_esperanza_vida(75))
print(clasificar_esperanza_vida(75.78))
print(clasificar_esperanza_vida(65))
print(clasificar_esperanza_vida(30.4))
print(clasificar_esperanza_vida(-19.5))

````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí
def obtener_media(serie:pd.Series)->float:
    suma = 0
    cantidad_datos = len(serie)
    for dato in serie:
        if pd.notna(dato):
            suma += dato
    return suma / cantidad_datos


lista_columnas_con_nulos = conteo_nulos[3:] # eliminamos columnas str
# separamos columnas con nulos
lista_columnas_con_nulos = [columna for columna, num_nulos in lista_columnas_con_nulos if num_nulos > 0]

for columa in lista_columnas_con_nulos:
    # media_columna = obtener_media(df_salud_mundial[columa]) # usando la funcion que construimos
    media_columna = df_salud_mundial[columa].median()
    df_salud_mundial[columa] = df_salud_mundial[columa].fillna(media_columna)

# df_salud_mundial.head(8)
print(df_salud_mundial.isnull().sum())

````
